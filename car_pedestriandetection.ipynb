```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img_file = "/content/car1.jfif"
img_file1="/content/Car.jpg"
# pre-trained car classifer
car_classifer = "/content/cars.xml"
car_classifer = "/content/cars.xml"

# Create opencv image( a multidimensional image data)
img = cv2.imread(img_file)
img1=cv2.imread(img_file1)


# Convert to greyscale
black_n_white = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
black_n_white = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# create  car classifer
car_tracker = cv2.CascadeClassifier(car_classifer)

# Detecting car in the image
cars = car_tracker.detectMultiScale(black_n_white)

# Drawing rectangle around the cars on the basis of the car cordinates
car1 = cars[0]
(x, y, w, h) = car1

cv2.rectangle(img, (x, y), (x+w, y+h), (0, 0, 255), 2)
cv2.rectangle(img1, (x, y), (x+w, y+h), (255, 0, 0), 2)

# for (x, y, w, h) in cars:
#     cv2.rectangle(img, (x, y), (x+w, y+h), (0, 0, 255), 2)


cv2_imshow(img)
cv2_imshow(img1)
cv2.waitKey(0)
```
